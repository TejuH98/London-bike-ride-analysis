```python
#Specifying column name that is prefered
new_col_nm = { 'timestamp' : 'time',
              'cnt' : 'count',
              't1' : 'temp_real_c',
              't2' : 'temp_feels_like_c',
              'hum' : 'humidity_perc',
              'wind_speed' : 'wind_speed_kph',
              'weather_code' : 'weather',
              'is_holiday' :'is_holiday',
              'is_weekend' :'is_weekend',
              'season ' : 'season'}
```

```python
#renaming the column to prefered column name
bike_df.rename(new_col_nm, inplace = True, axis = 1)
bike_df.head()
```

```python
#changing humidity percentage to actual percentage
bike_df.humidity_perc = bike_df.humidity_perc/100
```

```python
bike_df.head()
```

```python
#mapping code of weather and season to actual weather and season rather than its code
season_dict = { '0.0' : 'spring',
               '1.0' : 'summer',
               '2.0' : 'autumn',
               '3.0' : 'winter'}
weather_dict = {'1.0' : 'clear',
                '2.0' : 'scattered clouds',
                '3.0' : 'broken clouds',
                '4.0' : 'clouds',
                '7.0' : 'rain',
                '10.0' : 'rain with thunder',
                '26.0' : 'snowfall'}
```

```python
#changing data type and mapping string values to its integer codes
bike_df.season = bike_df.season.astype('str')
bike_df.weather = bike_df.weather.astype('str')

bike_df.season = bike_df.season.map(season_dict)
bike_df.weather = bike_df.weather.map(weather_dict)
```
